```python
# web scraping Tankathon's 2020 NBA Draft for player attributes and statistics 

import requests
# read webpage
r = requests.get('http://www.tankathon.com/big_board')
```

```python
print(r.text[0:500])
```

```python
from bs4 import BeautifulSoup
# read html
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
names_html = soup.find_all('div', attrs = {'class':'mock-row-name'})
names = []
for i in range(len(names_html)):
    names.append(names_html[i].text)
    print(names[i])
```

```python
school_position_html = soup.find_all('div', attrs = {'class': 'mock-row-school-position'})
positions = []
schools = []
for i in range(len(school_position_html)):
    line = str(school_position_html[i].text).split(' | ')
    positions.append(line[0])
    schools.append(line[1][0:-1])
    print("%s, %s" % (positions[i], schools[i]))
```

```python
height_weight_html = soup.find_all('div', attrs = {'class': 'section height-weight'})
heights = []
weights = []
for i in range(len(height_weight_html)):
    line = height_weight_html[i].contents
    height_split = line[0].text.split('\'')
    height = height_split[0]
    height += '\'\''
    height += height_split[1]
    
    weight_text = line[1].text
    weight_split = weight_text.split(' ')
    
    heights.append(height[0:-1])
    weights.append(weight_split[0])
    print("%s, %s" %(heights[i], weights[i]))
```

```python
year_age_html = soup.find_all('div', attrs = {'class': 'section year-age desktop'}) 
years = []
ages = []
for i in range(len(year_age_html)):
    line = year_age_html[i].contents
    
    years.append(line[0].text)
    age_split = line[1].text.split(' ')
    ages.append(age_split[0])
    
    # ages.append(line[1].text)
    print("%s, %s" %(years[i], ages[i]))
```

```python
# # print(scrapeStats("stat pts"))
# scrapeStats("stat pts")

search = soup.find_all('div', attrs = {'class': 'stats-per-game'})
pts = []
reb = []
ast = []
blk = []
stl = []
for i in range(len(search)+1):
    if i < 63:
        line = search[i].contents
        pts_line = line[0].contents
        reb_line = line[1].contents
        ast_line = line[2].contents
        blk_line = line[3].contents
        stl_line = line[4].contents

        pts.append(pts_line[0].text)
        reb.append(reb_line[0].text)
        ast.append(ast_line[0].text)
        blk.append(blk_line[0].text)
        stl.append(stl_line[0].text)
        
    elif i == 63:
        # handle jay scrubb
        pts.append('20.2')
        reb.append('8.9')
        ast.append('1.5')
        blk.append('1.6')
        stl.append('1.1')
    elif i > 63:
        line = search[i-1].contents
        pts_line = line[0].contents
        reb_line = line[1].contents
        ast_line = line[2].contents
        blk_line = line[3].contents
        stl_line = line[4].contents

        pts.append(pts_line[0].text)
        reb.append(reb_line[0].text)
        ast.append(ast_line[0].text)
        blk.append(blk_line[0].text)
        stl.append(stl_line[0].text)

print(len(pts))
print(len(reb))
print(len(ast))
print(len(blk))
print(len(stl))
```
