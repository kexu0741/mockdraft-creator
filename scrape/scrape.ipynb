```python
# web scraping Tankathon's 2020 NBA Draft for player attributes and statistics 

import requests
# read webpage
r = requests.get('http://www.tankathon.com/big_board')
```

```python
print(r.text[0:500])
```

```python
from bs4 import BeautifulSoup
# read html
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
names_html = soup.find_all('div', attrs = {'class':'mock-row-name'})
names = []
for i in range(len(names_html)):
    names.append(names_html[i].text)
    print(names[i])
```

```python
school_position_html = soup.find_all('div', attrs = {'class': 'mock-row-school-position'})
positions = []
schools = []
for i in range(len(school_position_html)):
    line = str(school_position_html[i].text).split(' | ')
    positions.append(line[0])
    schools.append(line[1])
    print("%s, %s" % (positions[i], schools[i]))
```

```python
height_weight_html = soup.find_all('div', attrs = {'class': 'section height-weight'})
heights = []
weights = []
for i in range(len(height_weight_html)):
    line = height_weight_html[i].contents
    height = line[0].text
    weight = line[1].text
    
    heights.append(height[0:-1])
    weights.append(weight)
    print("%s, %s" %(heights[i], weights[i]))
```

```python
year_age_html = soup.find_all('div', attrs = {'class': 'section year-age desktop'}) 
years = []
ages = []
for i in range(len(year_age_html)):
    line = year_age_html[i].contents
    
    years.append(line[0].text)
    ages.append(line[1].text)
    print("%s, %s" %(years[i], ages[i]))
```

```python
def scrapeStats(class_name):
    search = soup.find_all('div', attrs = {'class': class_name})
    
    ret = []
    for i in range(len(search)):
        line = search[i].contents
        classes = line[0].get('class')
        if 'stat-quality' not in classes:
            # todo: handle jay scrubb somehow
            if line[-1].text == 'pts':
                print(line)
                ret.append(line[0].text)
    print(len(ret))
    return ret
```
